```python
import matplotlib.pyplot as plt
%matplotlib inline
from Spectral_Line_Shape import *
import os
import csv
import numpy as np
import warnings
import random
import operator
import pandas as pd
```

```python
x_axis_total = []
y_axis_total = []

x_axis = []
solar_spectrum = []
with open("SolarSpectrum.csv") as csvfile :
    data = csv.reader(csvfile , delimiter = ',')
    
    for row in data :
        try :
            x_axis.append(float(row[0]))
            solar_spectrum.append(float(row[2])*10000)
        except :
            pass

for filename in os.listdir() :
    if filename[-3:] == 'txt':
        intensity , x_axis_w = Textfile_read(filename)
        intensity_normalized = np.array(intensity)/max(intensity)
        peak = list(intensity_normalized).index(max(intensity_normalized))
        param , x , p_0 , w = spectral_line_fit(x_axis_w , intensity_normalized , peak)
        x_new = (p_0 - np.array(x_axis))/(w/2)
        RMSE = param[0]
        n = param[1]
        Voigt = (n*Gaussian(x_new) + (1-n)*Lorentz(x_new))*max(intensity)
        plt.figure()
        plt.plot(x_axis_w , intensity , 'y' , label = 'LED Spectrum for  %s LED'  % filename[:5])
        plt.plot(x_axis , Lorentz(x_new) , 'r--' , label = 'Lorentz Model')
        plt.plot(x_axis , Gaussian(x_new) , 'b--' , label = 'Gaussian Model')
        plt.plot(x_axis , Voigt , 'g' , label = 'Combination with n = %1.2f , Error = %1.3f%% ' % (n , RMSE*100))
        plt.grid()
        plt.title("Speactral Line Shape has Peak Wavelength %3.2f nm" % x_axis_w[peak])
        plt.legend(loc='lower center', bbox_to_anchor=(1.4, 0.4))
        plt.xlim(300 , 1200)
        
        x_axis_total.append(x_axis)
        y_axis_total.append(Voigt)
```

```python
def RMSE(absError , yData) : 
    SE = np.square(absError) # squared errors
    MSE = np.mean(SE) # mean squared errors
    RMSE = np.sqrt(MSE) # Root Mean Squared Error, RMSE
    
    return RMSE
```

```python
def weighted_sum(y_axis_total , weights) :
    
    weighted_output = []
    for i in range(len(y_axis_total)) :
        weighted_output.append(weights[i]*y_axis_total[i])
        
    return np.array(weighted_output)
     
        
```

```python
class Params : 
    
    def __init__(self , weights) :
        self.weights = np.array(weights)
        
    def distance(self) :
        spectrum = weighted_sum(y_axis_total , self.weights)
        
        absError = abs(np.array(solar_spectrum) - np.array(spectrum))
        distance = RMSE(absError , np.array(solar_spectrum))
        
        return distance
    
    def __repr__(self):
        return "(" + str(self.weights) +  ")"
```

```python
def initialPopulation(popSize):
    population = []
    
    for i in range(0, popSize):
        population_temp = []
        for item in y_axis_total :
            if max(item) < 1000 :
                population_temp.append(random.uniform(0,100))
            elif max(item) > 7000 :
                population_temp.append(random.uniform(0,100))
            else :
                population_temp.append(random.uniform(0,100))
        population.append(Params(weights = population_temp))
        
    return population
```

```python
class Fitness :
    def __init__(self , spectrum) :
        self.spectrum = spectrum
        self.distance = 0
        self.fitness = 0.0     
        
    def spectrumdistance(self) :
        if self.distance == 0 :
            total_spectrum = self.spectrum
            pathdistance = total_spectrum.distance()
            
        self.distance = pathdistance
        
        return self.distance
       
    def spectrumfitness(self) :
        if self.fitness == 0 :
            self.fitness = 1 / float(self.spectrumdistance())
            
        return self.fitness
```
