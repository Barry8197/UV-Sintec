```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
f = open("400nm-38mA-100ms-2BC-Scans3-BG-1.txt" , 'r')
data = f.read()
```

```python
intensity = []
x_axis = []
for item in data.split('\n')[:-1] :
    intensity_temp , x_axis_temp = item.split(' ')
    x_axis.append(float(intensity_temp))
    intensity.append(float(x_axis_temp))
```

```python
plt.plot(x_axis , intensity)
```

```python
peak_orig = intensity.index(max(intensity))
```

```python
x_axis_cropped = x_axis[peak_orig - 300 : peak_orig + 300]
intensity_cropped = intensity[peak_orig - 300 : peak_orig + 300]
peak = intensity_cropped.index(max(intensity_cropped))
```

```python
plt.plot(x_axis_cropped , intensity_cropped)
```

```python
FWHM = [[1,1,1,1,1]]
rtol = 5
while len(FWHM[0]) > 4 :
    rtol = rtol/5
    fwhm = np.isclose(intensity_cropped , max(intensity_cropped)/2 , atol = .1 , rtol = rtol)
    FWHM = np.where(fwhm == True)
    
```

```python
def Gaussian(x) :
    x = np.array(x)
    Gauss = np.exp(-1*np.log(2)*(x**2))
    
    return Gauss
```
