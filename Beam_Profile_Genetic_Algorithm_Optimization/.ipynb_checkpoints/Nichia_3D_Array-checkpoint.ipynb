```python
angle = np.linspace(-70 , 70 , 181)
theta = angle*np.pi/180
axis_x = axis_y = np.tan(theta)
X_axis , Y_axis = np.meshgrid(axis_x , axis_y)

origin_x = np.linspace(-1,1,2)
origin_y = np.linspace(-1,1,2)
source_dis = np.linspace(0.1 , 2 , 10)
```

```python
def profile_adjustment(origin_x , origin_y , source_dis) :
    Z = []
    for i in range(len(X_axis)) :
        Z_temp = []
        for a , b in zip(X_axis[i] , Y_axis[i]) :
            dis_from_origin = (np.sqrt((a - origin_x)**2 + (b - origin_y)**2))
            angle_xy = np.arctan(dis_from_origin/(1+source_dis))
            Z_temp.append((I_calc_cos2([angle_xy] , c1 , c2 ,c3)[0])/(1+source_dis)**2)
        Z.append(Z_temp)
        
    return np.array(Z)
```

```python
variance = []
peak_intensity = []
for item in source_dis:
    Beam_Profile = np.array([np.zeros(181) for i in range(181)])
    for item1 in origin_x :
        for item2 in origin_y :
            Beam_Profile += np.array(profile_adjustment(item1 , item2 , item))
    
    max_val = max([max(item) for item in Beam_Profile])
    peak_intensity.append(max_val)
    fig = plt.figure(figsize = (15,6))
    ax1 = fig.add_subplot(121, projection='3d')
    ax1.plot_surface(X_axis, Y_axis, Beam_Profile, rstride=10, cstride=10 )
    ax1.plot_surface(X_axis, Y_axis, target, rstride=10, cstride=10 )
    ax1.view_init(10 , 30)
    ax2 = fig.add_subplot(122)
    cf = ax2.contourf(X_axis , Y_axis , Beam_Profile)
    cbar = plt.colorbar(cf)
    plt.show()
    absError = abs(target - Beam_Profile)
    Error = RMSE(absError , target)
    Error = Error*100
    print("RMSE Error is %2.2f %%" % Error)
    variance.append(np.var(Beam_Profile))
```

```python
def target() :
    Z = []
    for i in range(len(X_axis)) :
        Z_temp = []
        for a , b in zip(X_axis[i] , Y_axis[i]) :
            Z_temp.append(1)
        Z.append(Z_temp)
        
    return np.array(Z)
```

```python
class Params :
    
    def __init__(self , origin_x , origin_y , source_dis) :
        self.origin_x = origin_x
        self.origin_y = origin_y
        self.source_dis = source_dis
        
    def distance(self) :
        Beam_Profile = np.array([np.zeros(181) for i in range(181)])
        for item1 , item2 in zip(self.origin_x , self.origin_y) :
            Beam_Profile += profile_adjustment(item1 , item2 , self.source_dis)
            
        absError = abs(target - Beam_Profile)
        distance = RMSE(absError , target)
        
        return distance
    
    def __repr__(self):
        return "(" + str(self.origin_x) + "," + str(self.origin_y) +  "," + str(self.source_dis) + ")"
        
```

```python
import random
```

```python
def initialPopulation(popSize):
    population = []

    for i in range(0, popSize):
        source_dis = random.uniform(0,2)
        population.append(Params(origin_x= [random.uniform(-2 , 2) for i in range(4)] , 
                                 origin_y= [random.uniform(-2,2) for i in range(4)],
                                 source_dis= source_dis))
    return population
```
